```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from  dataPath import DATA_PATH
%matplotlib inline

import matplotlib
matplotlib.rcParams['image.cmap'] ="gray"
matplotlib.rcParams['figure.figsize'] = (6.0,6.0)
```

```python
im = cv.imread(DATA_PATH+"images/book.jpeg")
imGray = cv.cvtColor(im,cv.COLOR_BGR2GRAY)
plt.subplot(121);plt.imshow(im[...,::-1]);
plt.subplot(122);plt.imshow(imGray);
```

```python
orb = cv.ORB_create()
#find the keypoints with ORB
kp = orb.detect(imGray,None)
#compute the descriptors with ORB
kp,desc = orb.compute(imGray,kp)

#draw keypoints location, size and orientation
outImage = cv.drawKeypoints(im,kp,None, color = (255,240,0),flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(outImage[...,::-1])
```

```python
orb = cv.ORB_create()
kp, desc = orb.detectAndCompute(imGray,None)
outImage = cv.drawKeypoints(im,kp,None,color=(255,140,0),flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(outImage[...,::-1])
```
