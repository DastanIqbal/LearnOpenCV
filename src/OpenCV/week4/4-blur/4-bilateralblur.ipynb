```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,20)
matplotlib.rcParams["image.cmap"]="gray"
matplotlib.rcParams["axes.titlesize"] = 20
matplotlib.rcParams['axes.titlecolor']="white"

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"

image = cv2.imread(DATA_PATH+"images/gaussian-noise.png")
```

```python
diameter = 15

# Larger the value the distant colours will be mixed together
# to produce areas of semi equal colors
sigmaColor = 80

# Larger the value more the influence of the farther placed pixels
# as long as their colors are close enough
sigmaSpace = 80

result = cv2.bilateralFilter(image,diameter,sigmaColor,sigmaSpace)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(122);plt.imshow(result[...,::-1]);plt.title("Bi-Lateral Blur")
```
