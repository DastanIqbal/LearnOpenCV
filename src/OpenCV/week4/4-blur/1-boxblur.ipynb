```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,20)
matplotlib.rcParams["image.cmap"]="gray"
matplotlib.rcParams["axes.titlesize"] = 20
matplotlib.rcParams['axes.titlecolor']="white"

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"

image = cv2.imread(DATA_PATH+"images/gaussian-noise.png")
```

```python
blur3 = cv2.blur(image,(3,3),(-1,-1))
blur7 = cv2.blur(image,(7,7),(-1,-1))
```

```python
plt.figure(figsize=(10,10))
plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Original Image")
plt.subplot(132);plt.imshow(blur3[...,::-1]);plt.title("Blur 3x3 Image")
plt.subplot(133);plt.imshow(blur7[...,::-1]);plt.title("Blur 7x7 Image")
```
