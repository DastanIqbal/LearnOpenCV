```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,20)
matplotlib.rcParams["image.cmap"]="gray"
matplotlib.rcParams["axes.titlesize"] = 20
matplotlib.rcParams['axes.titlecolor']="white"

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"

image = cv2.imread(DATA_PATH+"images/gaussian-noise.png")
```

```python
dst1 = cv2.GaussianBlur(image,(5,5),0,0)
dst2 = cv2.GaussianBlur(image,(25,25),50,50)

lineType = 4
fontScale = 1

combined = np.hstack([image,dst1,dst2])

plt.imshow(combined[...,::-1])
```

```python
plt.figure(figsize=(20,20))
plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(132);plt.imshow(dst1[...,::-1]);plt.title("GaussianBlur 5x5 Kernel")
plt.subplot(133);plt.imshow(dst2[...,::-1]);plt.title("GaussianBlur 25x25 Kernel")
```
