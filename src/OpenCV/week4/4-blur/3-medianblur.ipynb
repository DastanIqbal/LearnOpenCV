```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,20)
matplotlib.rcParams["image.cmap"]="gray"
matplotlib.rcParams["axes.titlesize"] = 20
matplotlib.rcParams['axes.titlecolor']="white"

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"

image = cv2.imread(DATA_PATH+"images/salt-and-pepper.png")
```

```python
kernelSize = 5

medianBlur = cv2.medianBlur(image,kernelSize)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original")
plt.subplot(122);plt.imshow(medianBlur[...,::-1]);plt.title("MedianBlur 5x5 Kernel")
```
