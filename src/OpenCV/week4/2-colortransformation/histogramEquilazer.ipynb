```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"

import matplotlib

matplotlib.rcParams["figure.figsize"] = (10, 10)
matplotlib.rcParams["image.interpolation"] = "bilinear"
matplotlib.rcParams["image.cmap"] = "gray"
matplotlib.rcParams["axes.titlesize"] = 20
matplotlib.rcParams['axes.titlecolor']="white"
```

```python
filename = DATA_PATH + "images/dark-flowers.jpg"

im = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)

imEq = cv2.equalizeHist(im)

ax = plt.subplot(121);
plt.imshow(im, vmin=0, vmax=255)
ax.set_title("Original Images");
ax.axis("off")

ax = plt.subplot(122);
plt.imshow(imEq, vmin=0, vmax=255)
ax.set_title("History Equalized");
ax.axis("off")
```

```python
plt.figure(figsize=(20,20))
plt.subplot(121)
plt.hist(im.ravel(),256,[0,256])
plt.show()

plt.subplot(122)
plt.hist(imEq.ravel(),256,[0,256])
plt.show()
```

Equilzer for Color Images

Wrong Way

```python
im = cv2.imread(filename)
imEq = np.zeros_like(im)

for i in range(0,3):
    imEq[:,:,i] = cv2.equalizeHist(im[:,:,i])

ax = plt.subplot(121)
plt.imshow(im[...,::-1],vmin=0,vmax=255)
ax.set_title("Original Image");ax.axis("off")

ax = plt.subplot(122)
plt.imshow(imEq[...,::-1],vmin=0,vmax=255)
ax.set_title("Histogram Equilazer");ax.axis("off")
```

Equilzer for Color Images

Right Way

```python
im = cv2.imread(filename)

imhsv = cv2.cvtColor(im,cv2.COLOR_BGR2HSV)

imhsv[:,:,2] = cv2.equalizeHist(im[:,:,2])

imEq = cv2.cvtColor(imhsv,cv2.COLOR_HSV2RGB)

ax = plt.subplot(121)
plt.imshow(im[...,::-1],vmin=0,vmax=255)
ax.set_title("Original Image");ax.axis("off")

ax = plt.subplot(122)
plt.imshow(imEq,vmin=0,vmax=255)
ax.set_title("Histogram Equilazer");ax.axis("off")
```
