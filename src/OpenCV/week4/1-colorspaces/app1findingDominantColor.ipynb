```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib

matplotlib.rcParams["figure.figsize"] = (10, 10)
matplotlib.rcParams["image.interpolation"] = "bilinear"

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"


def axisLabelColor(ax):
    ax.tick_params(axis='x', colors='white', labelsize=20)
    ax.tick_params(axis='y', colors='white', labelsize=20)
```

```python
filename = DATA_PATH + "images/jersey.jpg"

img = cv2.imread(filename)

plt.figure(figsize=[20, 10])
plt.imshow(img[..., ::-1])
plt.axis("off")
```

```python
hsvImage = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

H, S, V = cv2.split(hsvImage)
print(H)
print(H.shape)
```

```python
# Remove unsaturated (white/gray) pixels

H_array = H[S>10].flatten()
print(H_array)
print(H_array.shape)
```

```python
plt.figure(figsize=[20, 10])
plt.subplot(121);
plt.imshow(img[..., ::-1]);
plt.title("Image");
plt.axis("off")
axH = plt.subplot(122);
plt.hist(H_array, bins=180, color='r');
plt.title("Histogram")

axisLabelColor(axH)
```
