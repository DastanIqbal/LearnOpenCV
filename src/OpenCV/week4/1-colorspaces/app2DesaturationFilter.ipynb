```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib

matplotlib.rcParams["figure.figsize"] = (10, 10)
matplotlib.rcParams["image.interpolation"] = "bilinear"

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"


def axisLabelColor(ax):
    ax.tick_params(axis='x', colors='white', labelsize=20)
    ax.tick_params(axis='y', colors='white', labelsize=20)
```

```python
filename = DATA_PATH + "images/girl.jpg"

img = cv2.imread(filename)

plt.figure(figsize=[20, 10])
plt.imshow(img[..., ::-1])
plt.axis("off")
```

```python
saturationScale = 0.01

hsvImage = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hsvImage = np.float32(hsvImage)
H, S, V = cv2.split(hsvImage)
```

```python
# Multiply S channel by scaling factor and clip the values to stay in 0 to 255
S = np.clip(S * saturationScale, 0 ,255)

hsvImage = np.uint8(cv2.merge([H,S,V]))
imSat = cv2.cvtColor(hsvImage,cv2.COLOR_HSV2BGR)
```

```python
plt.figure(figsize=[20, 10])
plt.subplot(121);plt.imshow(img[..., ::-1]);plt.title("Original Image");plt.axis("off")
plt.subplot(122);plt.imshow(imSat[..., ::-1]);plt.title("Desaturated Image");plt.axis("off")
```
