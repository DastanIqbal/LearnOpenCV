```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import matplotlib

matplotlib.rcParams["figure.figsize"] = (20, 15)
matplotlib.rcParams["image.cmap"] = "gray"

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
hsv = cv.imread(DATA_PATH + "images/capsicum.jpg", cv.COLOR_BGR2HSV)
plt.imshow(hsv[..., ::-1])
```

```python
plt.subplot(131);
plt.imshow(hsv[..., 0], cmap="gray");
plt.title("Hue/Color")
plt.subplot(132);
plt.imshow(hsv[..., 1], cmap="gray");
plt.title("Saturation/Shading")
plt.subplot(133);
plt.imshow(hsv[..., 2], cmap="gray");
plt.title("Value/Brightness/Toning")
```

Hue Test

```python
plt.figure(figsize=(50, 50))

for i in range(0, 7):
    imhsv = np.zeros((50, 50,3), dtype=np.uint8)

    h = i * 30
    imhsv[:, :, :] = (h, 128, 128)

    imrgb = cv.cvtColor(imhsv, cv.COLOR_HSV2RGB)
    ax = plt.subplot(1, 7, i + 1)
    plt.imshow(imrgb)
    plt.axis("off")
    ax.set_title(f"H = {str(h)}", fontdict={"fontsize": 50, "fontweight": "medium"})
```

Saturation Test

```python
plt.figure(figsize=(50,50))

for i in range(0,7):
    imhsv = np.zeros((50,50,3),dtype=np.uint8)

    s =i * 40
    imhsv[:,:,:] = (0, s, 128)
    imrgb = cv.cvtColor(imhsv,cv.COLOR_HSV2RGB)
    ax = plt.subplot(1,7,i+1)
    plt.imshow(imrgb)
    plt.axis("off")
    ax.set_title(f"S={str(s)}",fontdict={"fontsize":50,"fontweight":"medium"})
```

Value test

```python
plt.figure(figsize=(50,50))

for i in range(0,7):
    imhsv = np.zeros((50,50,3),dtype=np.uint8)

    v = i * 40
    imhsv[:,:,:] = (0, 0, v)

    imrgb = cv.cvtColor(imhsv,cv.COLOR_HSV2RGB)
    ax = plt.subplot(1,7,i+1)
    plt.imshow(imrgb)
    plt.axis("off")
    ax.set_title(f"V ={str(v)}",fontdict={"fontsize":50,"fontweight":"medium"})
```
