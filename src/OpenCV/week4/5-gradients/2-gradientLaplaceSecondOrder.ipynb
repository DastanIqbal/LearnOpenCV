```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
im = cv2.imread(DATA_PATH + "images/sample.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
img1 = cv2.GaussianBlur(im,(3,3),0,0)
laplcian = cv2.Laplacian(img1, cv2.CV_32F, ksize=3,scale=1,delta=0)

plt.figure(figsize=(15, 15))
plt.subplot(121);
plt.imshow(laplcian)
```

```python
cv2.normalize(laplcian, laplcian, 0, 1, cv2.NORM_MINMAX, cv2.CV_32F, )

plt.figure(figsize=(15, 15))
plt.subplot(121);
plt.imshow(laplcian,cmap="gray")
```
