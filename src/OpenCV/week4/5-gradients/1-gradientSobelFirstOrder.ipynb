```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
im = cv2.imread(DATA_PATH + "images/truth.png", cv2.IMREAD_GRAYSCALE)
```

```python
sobelX = cv2.Sobel(im, cv2.CV_32F, 1, 0)
sobelY = cv2.Sobel(im, cv2.CV_32F, 0, 1)

plt.figure(figsize=(15, 15))
plt.subplot(121);
plt.imshow(sobelX)
plt.subplot(122);
plt.imshow(sobelY)
```

```python
cv2.normalize(sobelX, sobelX, 0, 1, cv2.NORM_MINMAX, cv2.CV_32F, )
cv2.normalize(sobelY, sobelX, 1, 0, cv2.NORM_MINMAX, cv2.CV_32F, )

plt.figure(figsize=(15, 15))
plt.subplot(121)
plt.imshow(sobelX,cmap="gray")
plt.subplot(122)
plt.imshow(sobelY,cmap="gray")
```
