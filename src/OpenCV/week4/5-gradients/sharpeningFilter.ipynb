```python

import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
im = cv2.imread(DATA_PATH+"images/sample.jpg")
```

```python
kernel = np.array([0,-1,0,
                   -1,5,-1,
                   0,-1,0],dtype='int')

sharpenIm = cv2.filter2D(im,-1,kernel)

plt.subplot(121);plt.imshow(im[...,::-1]);plt.title("Original Image")
plt.subplot(122);plt.imshow(sharpenIm[...,::-1]);plt.title("Sharpened Image")
```

```python
#𝐼𝑠=𝐼+𝛼(𝐼−𝐼𝑏)

imb = cv2.GaussianBlur(im,(3,3),0)
shaprenIm = im + 1*(im - imb)
plt.imshow(shaprenIm[...,::-1])
```
