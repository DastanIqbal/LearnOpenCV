```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,20)
matplotlib.rcParams["image.cmap"]="gray"
matplotlib.rcParams["axes.titlesize"] = 20
matplotlib.rcParams['axes.titlecolor']="white"

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
filename = DATA_PATH + "images/sample.jpg"
image = cv2.imread(filename)

if image is None:
    print("Couldn't not open or find the image")
```

```python
kernel_size = 5
kernel = np.ones((kernel_size,kernel_size), dtype=np.float32) / kernel_size**2

print(kernel)
```

```python
result = cv2.filter2D(image,-1,kernel,(-1,-1),delta=0,borderType=cv2.BORDER_DEFAULT)
plt.figure(figsize=(20,20))
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original Image")
plt.subplot(122);plt.imshow(result[...,::-1]);plt.title("Convolution Image")
```
