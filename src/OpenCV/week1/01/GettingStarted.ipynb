```python
#Color Image
imagePath = os.path.join(DATA_PATH,"images/musk.jpg")
img = cv2.imread(imagePath)
print("Image Dimension = {}".format(img.shape))
plt.imshow(img)
imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #convert bgr to rgb
plt.imshow(imgRGB)
plt.imshow(img[:,:,::-1]) #reverse the order of channel
```

```python
plt.figure(figsize=[20,5])
plt.subplot(131);plt.imshow(img[:,:,0]);plt.title("Blue Channel")
plt.subplot(132);plt.imshow(img[:,:,1]);plt.title("Green Channel")
plt.subplot(133);plt.imshow(img[:,:,2]);plt.title("Red Channel")
```

```python
b,g,r = cv2.split(img)
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(img[:,:,0]);plt.title("Blue Channel")
plt.subplot(142);plt.imshow(img[:,:,1]);plt.title("Green Channel")
plt.subplot(143);plt.imshow(img[:,:,2]);plt.title("Red Channel")

imgMerged = cv2.merge((b,g,r))
plt.subplot(144);plt.imshow(imgMerged[:,:,::-1]);plt.title("Merged Output")
```

```python
# Images with Alpha Channel
imagePath = os.path.join(DATA_PATH,"images/panther.png")

imgPNG = cv2.imread(imagePath,cv2.IMREAD_UNCHANGED)
imgRGB = cv2.cvtColor(imgPNG,cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
print("Image Dimension = {}".format(imgPNG.shape))

imgBGR = imgPNG[:,:,0:3] #fetch 3 channels from PNG image
imgMask = imgPNG[:,:,3] #fetch alpha channel from PNG image

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(imgBGR[:,:,::-1]);plt.title("Color Channels")
plt.subplot(122);plt.imshow(imgMask,cmap="gray");plt.title("Alpha Channels")
```
