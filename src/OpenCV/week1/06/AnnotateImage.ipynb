```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline

import matplotlib as mat
mat.rcParams['figure.figsize'] = (6.0,6.0)
mat.rcParams['image.cmap'] = 'gray'

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
faceImage = cv2.imread(DATA_PATH+"images/boy.jpg")
faceImage = cv2.cvtColor(faceImage,cv2.COLOR_BGR2RGB)
plt.imshow(faceImage)
```

```python
# Line

imageLine = faceImage.copy()

cv2.line(imageLine,(200,80),(280,80),color=(0,255,0),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(imageLine)
```

```python
# Circle

imageCircle = faceImage.copy()
cv2.circle(imageCircle,(250,125),radius=100,color=(0,0,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(imageCircle)
```

```python
# Filled Circle

imageFilledCircle = faceImage.copy()
cv2.circle(imageFilledCircle,(250,125),radius=100,color=(0,0,255),thickness=-1,lineType=cv2.LINE_AA)
plt.imshow(imageFilledCircle)
```

```python
# Ellipse

imageEllipse = faceImage.copy()
cv2.ellipse(imageEllipse,(250,125),(100,50),0,0,360,(0,0,255),thickness=3,lineType=cv2.LINE_AA)
cv2.ellipse(imageEllipse,(250,125),(100,50),90,0,360,(255,0,0),thickness=3,lineType=cv2.LINE_AA)
cv2.ellipse(imageEllipse,(250,125),(100,50),45,0,360,(0,255,0),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(imageEllipse)
```

```python
# Filled Ellipse
imageEllipse = faceImage.copy()
cv2.ellipse(imageEllipse,(250,125),(100,50),0,0,180,(0,0,255),thickness=-3,lineType=cv2.LINE_AA)
cv2.ellipse(imageEllipse,(250,125),(100,50),0,180,360,(255,0,0),thickness=-3,lineType=cv2.LINE_AA)
plt.imshow(imageEllipse)
```

```python
# Draw a Rectangle

imageRectangle = faceImage.copy()
cv2.rectangle(imageRectangle,(150,30),(340,210),color=(0,255,0),thickness=4,lineType=cv2.LINE_AA)
plt.imshow(imageRectangle)
```

```python
# Add a Text

imageText = faceImage.copy()
text = "Hello CV"
fontScale = 1.5
fontFace = cv2.FONT_HERSHEY_COMPLEX
fontColor = (250, 10, 10)
fontThickness = 2

cv2.putText(imageText,text,(10,330),fontFace,fontScale,fontColor,fontThickness,cv2.LINE_AA)
plt.imshow(imageText)
```

```python
# Fixing text issues

imageText = faceImage.copy()
pixelHeight = 20
fontScale = cv2.getFontScaleFromHeight(fontFace,pixelHeight,fontThickness)
print("Font scale {}".format(fontScale))

cv2.putText(imageText,text,(20,350),fontFace,fontScale,fontColor,fontThickness,cv2.LINE_AA)
plt.imshow(imageText)
```

```python
imageGetTextSize = faceImage.copy()
imageHeight, imageWidth=imageGetTextSize.shape[:2]

# Get the text box height and width and also the baseLine
textSize, baseLine = cv2.getTextSize(text,fontFace,fontScale,fontThickness)
textWidth,textHeight = textSize
print("TextWidth = {}, TextHeight = {}, baseLine = {}".format(textWidth, textHeight, baseLine))

# Get the coordinates of text box bottom left corner
# The xccordinate will be such that the text is centered
xcoordinate = (imageWidth - textWidth)//2
# The y coordinate will be such that the entire box is just 10 pixels above the bottom of image
ycoordinate = (imageHeight - baseLine - 10)
print("TextBox Bottom Left = ({},{})".format(xcoordinate,ycoordinate))

# Draw the Canvas using a filled rectangle
canvasColor = (255, 255, 255)
canvasBottomLeft = (xcoordinate,ycoordinate+baseLine)
canvasTopRight = (xcoordinate+textWidth, ycoordinate-textHeight)
cv2.rectangle(imageGetTextSize, canvasBottomLeft, canvasTopRight, canvasColor, thickness=-1);
print("Canvas Bottom Left = {}, Top Right = {}".format(canvasBottomLeft,canvasTopRight))

# Now draw the baseline ( just for reference )
lineThickness = 2
lineLeft = (xcoordinate, ycoordinate)
lineRight = (xcoordinate+textWidth, ycoordinate)
lineColor = (0,255,0)
cv2.line(imageGetTextSize, lineLeft, lineRight, lineColor, thickness = lineThickness, lineType=cv2.LINE_AA);

# Finally Draw the text
cv2.putText(imageGetTextSize, text, (xcoordinate,ycoordinate), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA);

# Display the Output Image
plt.imshow(imageGetTextSize)
```
