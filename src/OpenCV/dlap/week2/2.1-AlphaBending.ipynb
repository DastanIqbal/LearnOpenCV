## Alpha Bending

Forumla: ùêº = ùõºùêπ + (1‚àíùõº)ùêµ


```python
import cv2,dlib
import numpy as np
from dataPath import DATA_PATH
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams['figure.figsize'] = (6.0,6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
foregroundImage = cv2.imread(DATA_PATH+"images/foreGroundAssetLarge.png",cv2.IMREAD_UNCHANGED)
(b,g,r,a) = cv2.split(foregroundImage)
```

```python
plt.imshow(foregroundImage[:,:,[2,1,0,3]])
plt.title("Foreground Image")
plt.show()
```

```python
foreground = cv2.merge((b,g,r))
alpha = cv2.merge((a,a,a))
background = cv2.imread(DATA_PATH+"images/backGroundLarge.jpg")
```

```python
plt.imshow(background[:,:,::-1])
plt.title("Background")
plt.show()
```

```python
# Convert uint8 to float
foreground = foreground.astype(float)
background = background.astype(float)
alpha = alpha.astype(float)/255
```

```python
# Perform alpha blending Forumla: ùêº = ùõºùêπ + (1‚àíùõº)ùêµ
foreground = cv2.multiply(alpha, foreground)
background = cv2.multiply(1.0 - alpha, background)
outImage = cv2.add(foreground, background)
```

```python
plt.imshow(outImage[:,:,::-1]/255)
plt.title("Alpha Bended Image")
plt.show()
```
