```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from dataPath import DATA_PATH
%matplotlib inline

import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,20)
matplotlib.rcParams["image.cmap"]="gray"
```

```python
src = cv2.imread(DATA_PATH + "images/airplane.jpg")
dst = cv2.imread(DATA_PATH + "images/sky.jpg")
```

```python
scr_mask = np.zeros(src.shape,src.dtype)
poly = np.array([[4,80],[30,54],[151,63],[254,37],[298,90],[272,134],[43,122]],np.int32)
src_mask = cv2.fillPoly(scr_mask,[poly],(255,255,255))
# plt.imshow(src_mask[...,::-1])
```

```python
center = (800,100)
```

```python
output = cv2.seamlessClone(src,dst,src_mask,center,cv2.NORMAL_CLONE)
```

```python
plt.imshow(output[...,::-1])
plt.show()
```

```python
src = cv2.imread(DATA_PATH + "images/iloveyouticket.jpg")
dst = cv2.imread(DATA_PATH + "images/wood-texture.jpg")

mask = 255 * np.ones(src.shape,src.dtype)
width, height, channels = dst.shape
center = (height//2, width//2)

normal_clone = cv2.seamlessClone(src,dst,mask,center,cv2.NORMAL_CLONE)
mixed_clone = cv2.seamlessClone(src,dst,mask,center,cv2.MIXED_CLONE)
```

```python
plt.figure(figsize=[20,10])
plt.subplot(121);plt.title("Normal Clone");plt.imshow(normal_clone[...,::-1])
plt.subplot(122);plt.title("Mixed Clone");plt.imshow(mixed_clone[...,::-1])
```
