```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams['figure.figsize']=(6,6)
matplotlib.rcParams['image.cmap']='gray'

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
im = cv2.imread(DATA_PATH +"images/opening.png",cv2.IMREAD_GRAYSCALE)

if im is None:
    print("Couldn't open or find the image")

plt.imshow(im)
```

```python
# Method 1
#Remove small white blobs while keeping the bigger blobs intact
kSize =10

#Create the Kernel
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2*kSize+1,2*kSize+1),(kSize,kSize))

plt.imshow(element)
```

```python
imEroded = cv2.erode(im,element,iterations=1)
imOpen = cv2.dilate(imEroded,element,iterations=2)

plt.subplot(131);plt.imshow(im);plt.title("Original")
plt.subplot(132);plt.imshow(imEroded,cmap='gray');plt.title("After Erosion")
plt.subplot(133);plt.imshow(imOpen,cmap='gray');plt.title("After Dilation")
```

```python
# Method 2

openingSize =3
element = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2*openingSize+1,2*openingSize+1),(openingSize,openingSize))

imMorphed = cv2.morphologyEx(im,cv2.MORPH_OPEN,element,iterations=3)

plt.figure(figsize=[10,10])
plt.subplot(121);plt.imshow(im);plt.title("Original")
plt.subplot(122);plt.imshow(imMorphed,cmap='gray');plt.title("After Opening Ops")
```
