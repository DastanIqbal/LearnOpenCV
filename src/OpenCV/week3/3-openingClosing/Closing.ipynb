```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
im = cv2.imread(DATA_PATH+"images/closing.png",cv2.IMREAD_GRAYSCALE)
if im is None:
    print("Couldn't open or find the image")

plt.imshow(im)
```

```python
#Method 1
kSize = 10
elements = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2*kSize+1,2*kSize+1),(kSize,kSize))

imDilate = cv2.dilate(im,elements,iterations=1)
imCloseing = cv2.erode(imDilate,elements,iterations=1)

plt.figure(figsize=[10,10])
plt.subplot(131);plt.imshow(im);plt.title("Original")
plt.subplot(132);plt.imshow(imDilate,cmap='gray');plt.title("After Dilation")
plt.subplot(133);plt.imshow(imCloseing,cmap='gray');plt.title("After Erossion")
```

```python
# Method 2
kSize = 10
elements = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(2*kSize+1,2*kSize+1),(kSize,kSize))

imMorphed = cv2.morphologyEx(im,cv2.MORPH_CLOSE,elements,iterations=2)

plt.figure(figsize=[10,10])
plt.subplot(131);plt.imshow(im[...,::-1]);plt.title("Original")
plt.subplot(132);plt.imshow(imMorphed,cmap='gray');plt.title("Morphed Closing")
```
