```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
im = cv2.imread(DATA_PATH+"images/dilation_example.jpg")

if im is None:
    print("Could not open or find the image")
plt.imshow(im)
```

```python
kSize = (7,7)
kernel1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,kSize)
plt.imshow(kernel1)
```

```python
imDilated = cv2.dilate(im,kernel1)

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(im);plt.title("Original")
plt.subplot(122);plt.imshow(imDilated);plt.title("Dilated")
```

```python
kSize=(3,3)
kernel2 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,kSize)
plt.imshow(kernel2)
```

```python
imDilatedSmall1 = cv2.dilate(im,kernel2,iterations=1)
imDilatedSmall2 = cv2.dilate(im,kernel2,iterations=2)

plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(im);plt.title("Original")
plt.subplot(132);plt.imshow(imDilatedSmall1);plt.title("Dilated Small I1")
plt.subplot(133);plt.imshow(imDilatedSmall2);plt.title("Dilated Small I2")
```
