```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"
```

```python
im = cv2.imread(DATA_PATH+"images/erosion_example.jpg")

if im is None:
    print("Could not open or find the image")
plt.imshow(im)
```

```python
kSize = (7,7)
kernel1 = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,kSize)
plt.imshow(kernel1)
```

```python
imErode=cv2.erode(im,kernel1)

plt.figure(figsize=[20,20])
plt.subplot(121);plt.imshow(im);plt.title("Original")
plt.subplot(122);plt.imshow(imErode);plt.title("Erode")
```
