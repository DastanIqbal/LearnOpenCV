```python
import cv2
%matplotlib inline

DATA_PATH = "/Users/admin/Iqbal/Study/OpenCV/LearnOpenCV/src/res/opencv-courses/"

import matplotlib

matplotlib.rcParams["figure.figsize"] = (6, 6)
matplotlib.rcParams["image.cmap"] = "gray"
```

```python
def resizeImage(action, x, y, flag, userdata):
    # Referencing global variables
    if action == cv2.EVENT_MOUSEWHEEL and (flag & cv2.EVENT_FLAG_CTRLKEY):
        # Action to be taken when ctrl key + mouse wheel scrolled forward
        if flag > 0:
            # Resize image
            print("Resize Image Up")
            # Action to be taken when ctrl key + mouse wheel scrolled backward
        else:
            print("Resize Image Down")
```

```python
im = cv2.imread(DATA_PATH + "images/musk.jpg")
cv2.imshow("scaleUp", im)

cv2.setMouseCallback("scaleUp", resizeImage)
```
